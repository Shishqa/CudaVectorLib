```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='darkgrid', font_scale=1.4)
```

```python
results_path = "03-matrix-add"

data = pd.read_csv("data/" + results_path + ".csv")
data.head()

data['time'] = data['time'] / 1000

#data = data[data['mode'] == 'r']
#data.head()
```

```python
def plot_slice(data, block_sz, ax):
    data_slice = data.loc[(data['arr'] == block_sz)]
    data_slice.plot(x = 'block', y = 'time', label="matrix side: {}".format(block_sz), ax=ax)
     
fig, ax = plt.subplots(figsize=(12, 7))

for block_sz in data['arr'].unique():
    plot_slice(data, block_sz, ax)

ax.set_yscale('log')
ax.set_title(results_path)
ax.set_ylabel('log(time)')
ax.set_xlabel('block size')
```

```python
fig.savefig("data/" + results_path + ".png")
```
